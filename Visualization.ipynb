0. Setup and Imports
Run this first to import the necessary libraries and set a clean visual theme.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean aesthetic for all plots
sns.set_theme(style="whitegrid")
np.random.seed(42) # For reproducible synthetic data
```

1. Histograms
Histograms are used to visualize the distribution of continuous data.

```python
# --- Synthetic Data ---
data1 = np.random.normal(loc=50, scale=10, size=1000)
data2 = np.random.normal(loc=70, scale=8, size=1000)

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('1. Histograms (5 Variations)', fontsize=16)

# 1. Basic Histogram
axes[0, 0].hist(data1, bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title('1. Basic Histogram')

# 2. Custom Bins and Colors
axes[0, 1].hist(data1, bins=np.arange(20, 90, 5), color='coral', edgecolor='white')
axes[0, 1].set_title('2. Custom Bins (Width=5)')

# 3. Overlaid Histograms (Comparing distributions)
axes[1, 0].hist(data1, bins=30, alpha=0.5, label='Group A', color='blue')
axes[1, 0].hist(data2, bins=30, alpha=0.5, label='Group B', color='red')
axes[1, 0].legend()
axes[1, 0].set_title('3. Overlaid Histograms')

# 4. Density Plot (KDE) with Histogram using Seaborn
sns.histplot(data1, kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('4. Histogram + KDE Curve')

# 5. Cumulative Histogram
axes[2, 0].hist(data1, bins=30, cumulative=True, color='green', edgecolor='black', alpha=0.7)
axes[2, 0].set_title('5. Cumulative Step Histogram')

# Remove empty subplot
fig.delaxes(axes[2,1])
plt.tight_layout()
plt.show()
```

2. Scatter Plots
Scatter plots visualize the relationship between two numerical variables.

```python
# --- Synthetic Data ---
x = np.random.rand(100) * 100
y = x * 1.5 + np.random.normal(0, 15, 100)
colors = np.random.rand(100)
sizes = np.random.rand(100) * 500

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('2. Scatter Plots (5 Variations)', fontsize=16)

# 1. Basic Scatter Plot
axes[0, 0].scatter(x, y, color='blue', alpha=0.7)
axes[0, 0].set_title('1. Basic Scatter')

# 2. Bubble Chart (Varying Sizes)
axes[0, 1].scatter(x, y, s=sizes, color='green', alpha=0.5, edgecolor='black')
axes[0, 1].set_title('2. Bubble Chart (Varying Size)')

# 3. Color Mapped Scatter (Using a Colormap)
scatter3 = axes[1, 0].scatter(x, y, c=colors, cmap='viridis', s=80)
fig.colorbar(scatter3, ax=axes[1, 0])
axes[1, 0].set_title('3. Color Mapped by Variable')

# 4. Different Markers
axes[1, 1].scatter(x[:50], y[:50], marker='^', color='red', label='Type 1')
axes[1, 1].scatter(x[50:], y[50:], marker='s', color='blue', label='Type 2')
axes[1, 1].legend()
axes[1, 1].set_title('4. Distinct Markers for Categories')

# 5. Scatter Plot with Regression Line
sns.regplot(x=x, y=y, ax=axes[2, 0], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[2, 0].set_title('5. Scatter + Regression Line')

fig.delaxes(axes[2,1])
plt.tight_layout()
plt.show()
```

3. Bar Plots
Bar plots are best for comparing categorical data.

```python
# --- Synthetic Data ---
categories = ['Q1', 'Q2', 'Q3', 'Q4']
values_A = [25, 40, 35, 50]
values_B = [20, 35, 30, 45]
errors = [2, 3, 2.5, 4]

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('3. Bar Plots (5 Variations)', fontsize=16)

# 1. Basic Vertical Bar
axes[0, 0].bar(categories, values_A, color='teal')
axes[0, 0].set_title('1. Basic Vertical Bar')

# 2. Horizontal Bar Plot
axes[0, 1].barh(categories, values_A, color='orange')
axes[0, 1].set_title('2. Horizontal Bar')

# 3. Grouped Bar Plot
x_indexes = np.arange(len(categories))
width = 0.35
axes[1, 0].bar(x_indexes - width/2, values_A, width=width, label='Product A', color='blue')
axes[1, 0].bar(x_indexes + width/2, values_B, width=width, label='Product B', color='cyan')
axes[1, 0].set_xticks(x_indexes)
axes[1, 0].set_xticklabels(categories)
axes[1, 0].legend()
axes[1, 0].set_title('3. Grouped Bar Plot')

# 4. Stacked Bar Plot
axes[1, 1].bar(categories, values_A, label='Product A', color='purple')
axes[1, 1].bar(categories, values_B, bottom=values_A, label='Product B', color='pink')
axes[1, 1].legend()
axes[1, 1].set_title('4. Stacked Bar Plot')

# 5. Bar Plot with Error Bars
axes[2, 0].bar(categories, values_A, yerr=errors, capsize=5, color='gray', edgecolor='black')
axes[2, 0].set_title('5. Bar Plot with Error Bars')

fig.delaxes(axes[2,1])
plt.tight_layout()
plt.show()
```

4. Box Plots
Box plots display the summary of a set of data values having properties like minimum, first quartile, median, third quartile and maximum.

```python
# --- Synthetic Data ---
np.random.seed(10)
data_group1 = np.random.normal(100, 10, 200)
data_group2 = np.random.normal(90, 20, 200)
data_group3 = np.random.normal(105, 5, 200)
data_list = [data_group1, data_group2, data_group3]
labels = ['Group X', 'Group Y', 'Group Z']

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('4. Box Plots (5 Variations)', fontsize=16)

# 1. Single Basic Box Plot
axes[0, 0].boxplot(data_group1, labels=['Group X'])
axes[0, 0].set_title('1. Single Box Plot')

# 2. Multiple Box Plots
axes[0, 1].boxplot(data_list, labels=labels)
axes[0, 1].set_title('2. Multiple Box Plots')

# 3. Horizontal Box Plot
axes[1, 0].boxplot(data_list, labels=labels, vert=False)
axes[1, 0].set_title('3. Horizontal Box Plots')

# 4. Customized/Notched Box Plot
axes[1, 1].boxplot(data_list, labels=labels, notch=True, patch_artist=True,
                   boxprops=dict(facecolor='lightblue', color='blue'),
                   medianprops=dict(color='red', linewidth=2))
axes[1, 1].set_title('4. Notched & Styled Box Plot')

# 5. Box Plot with Swarm Overlay (Seaborn)
df_box = pd.DataFrame({'Value': np.concatenate(data_list),
                       'Group': np.repeat(labels, 200)})
sns.boxplot(x='Group', y='Value', data=df_box, ax=axes[2, 0], color='lightgray')
sns.stripplot(x='Group', y='Value', data=df_box, ax=axes[2, 0], color='black', alpha=0.3, size=3)
axes[2, 0].set_title('5. Box Plot + Strip/Swarm Plot')

fig.delaxes(axes[2,1])
plt.tight_layout()
plt.show()
```

5. Line Plots & Curves
Line plots connect individual data points. They are generally used for time series or mathematical curves.

```python
# --- Synthetic Data ---
x_time = np.arange(0, 10, 0.5)
y_trend1 = x_time * 2 + np.random.normal(0, 2, len(x_time))
y_trend2 = x_time * 3 + np.random.normal(0, 3, len(x_time))

# Math curve data
x_math = np.linspace(-10, 10, 400)
y_math = np.sin(x_math) * np.exp(-0.1 * x_math)

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('5. Line Plots & Curves (5 Variations)', fontsize=16)

# 1. Basic Line Plot
axes[0, 0].plot(x_time, y_trend1, marker='o', color='blue')
axes[0, 0].set_title('1. Basic Line Plot with Markers')

# 2. Multiple Lines with Styles
axes[0, 1].plot(x_time, y_trend1, linestyle='--', color='red', label='Metric A')
axes[0, 1].plot(x_time, y_trend2, linestyle='-.', color='green', label='Metric B')
axes[0, 1].legend()
axes[0, 1].set_title('2. Multiple Lines & Styles')

# 3. Mathematical Curve ($y = \sin(x) e^{-0.1x}$)
axes[1, 0].plot(x_math, y_math, color='purple', linewidth=2)
axes[1, 0].set_title('3. Mathematical Curve')

# 4. Parametric Curve (e.g., a Spiral)
t = np.linspace(0, 4 * np.pi, 1000)
x_para = t * np.cos(t)
y_para = t * np.sin(t)
axes[1, 1].plot(x_para, y_para, color='orange')
axes[1, 1].set_title('4. Parametric Curve (Spiral)')

# 5. Filled Area Under Line
axes[2, 0].plot(x_time, y_trend1, color='navy')
axes[2, 0].fill_between(x_time, y_trend1, alpha=0.3, color='navy')
axes[2, 0].set_title('5. Area Chart (Filled Area)')

fig.delaxes(axes[2,1])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional, clean theme for business reporting
sns.set_theme(style="whitegrid", palette="muted")
np.random.seed(42) # Ensures the random data looks the same every time you run it
```

1. Histograms: Understanding Business Distributions
Use these to find bottlenecks, demographic peaks, and behavior patterns.

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 15))
fig.suptitle('1. Histograms: Business Distributions', fontsize=18, fontweight='bold')

# 1. Order Processing Time (Right-skewed: Most are fast, some take very long)
processing_times = np.random.lognormal(mean=1.5, sigma=0.8, size=1000)
axes[0, 0].hist(processing_times, bins=40, range=(0, 30), color='steelblue', edgecolor='black')
axes[0, 0].set(title='1. Order Processing Time', xlabel='Hours to Process', ylabel='Number of Orders')
axes[0, 0].axvline(np.median(processing_times), color='red', linestyle='dashed', linewidth=2, label=f'Median: {np.median(processing_times):.1f}h')
axes[0, 0].legend()

# 2. Customer Age Demographics (Normal distribution around a target age)
ages = np.random.normal(loc=32, scale=8, size=1000)
axes[0, 1].hist(ages, bins=25, color='mediumseagreen', edgecolor='black')
axes[0, 1].set(title='2. Customer Age Demographics', xlabel='Age', ylabel='Customer Count')

# 3. Shopping Cart Value (Bimodal/Skewed: Small standard orders + fewer large orders)
cart_values = np.concatenate([np.random.normal(30, 10, 700), np.random.normal(120, 30, 300)])
axes[1, 0].hist(cart_values, bins=30, color='coral', edgecolor='black')
axes[1, 0].set(title='3. Shopping Cart Value Distribution', xlabel='Order Value ($)', ylabel='Frequency')

# 4. Employee Tenure (Peak at 1 year, tapering off)
tenure_months = np.random.gamma(shape=2, scale=12, size=500)
axes[1, 1].hist(tenure_months, bins=24, color='purple', alpha=0.7, edgecolor='white')
axes[1, 1].set(title='4. Employee Tenure at Exit', xlabel='Months of Employment', ylabel='Number of Employees')

# 5. Website Session Duration (Exponential decay: high bounce rate)
session_seconds = np.random.exponential(scale=45, size=2000)
axes[2, 0].hist(session_seconds, bins=50, range=(0, 300), color='goldenrod', edgecolor='black')
axes[2, 0].set(title='5. Web Session Duration (Bounce Rate View)', xlabel='Seconds on Site', ylabel='Number of Sessions')

fig.delaxes(axes[2, 1]) # Remove empty subplot
plt.tight_layout()
plt.show()
```

**Scatter Plots (Finding Relationships) - Use Cases**

Marketing Ad Spend vs. Revenue Generation: Plotting monthly spend on the X-axis and resulting sales on the Y-axis.

Business Value: Measures ROI. It shows if spending more actually yields linear returns, or if there is a point of diminishing returns.



---



Product Price vs. Units Sold: Mapping different price points against sales volume across different regions or times.

Business Value: Determines price elasticity. Helps find the optimal price point that maximizes total profit, not just unit volume.



---

Business Value: Sets service level agreements (SLAs). It reveals exactly how many hours you have to respond before customer satisfaction plummets.

Employee Salary vs. Years of Experience: Plotting compensation data.

Business Value: Supply chain optimization. Helps determine if establishing a new micro-fulfillment center in a specific region would be cost-effective.

2. Scatter Plots: Finding Business Relationships
Use these to prove ROI, find price elasticity, and set operational KPIs.

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 15))
fig.suptitle('2. Scatter Plots: Business Relationships', fontsize=18, fontweight='bold')

# 1. Ad Spend vs. Revenue (Positive correlation with diminishing returns)
spend = np.random.uniform(1000, 10000, 100)
revenue = spend * np.log(spend) * 0.5 + np.random.normal(0, 5000, 100)
sns.regplot(x=spend, y=revenue, ax=axes[0, 0], color='blue', scatter_kws={'alpha':0.6})
axes[0, 0].set(title='1. Marketing ROI: Ad Spend vs Revenue', xlabel='Ad Spend ($)', ylabel='Revenue Generated ($)')

# 2. Price vs. Units Sold (Negative correlation - Price Elasticity)
prices = np.random.uniform(10, 100, 80)
units_sold = 1000 - (prices * 8) + np.random.normal(0, 50, 80)
axes[0, 1].scatter(prices, units_sold, color='crimson', alpha=0.7)
axes[0, 1].set(title='2. Price Elasticity: Price vs. Units Sold', xlabel='Product Price ($)', ylabel='Units Sold')

# 3. CSAT vs. Support Response Time (Negative correlation)
response_hours = np.random.uniform(0.5, 48, 150)
csat_score = 100 - (response_hours * 1.5) + np.random.normal(0, 5, 150)
csat_score = np.clip(csat_score, 0, 100) # Keep scores between 0-100
scatter = axes[1, 0].scatter(response_hours, csat_score, c=csat_score, cmap='RdYlGn', alpha=0.8)
axes[1, 0].set(title='3. SLA Impact: Response Time vs. CSAT', xlabel='Response Time (Hours)', ylabel='CSAT Score (0-100)')
fig.colorbar(scatter, ax=axes[1, 0], label='Satisfaction Level')

# 4. Salary vs. Experience (Positive correlation, variance widens with experience)
experience = np.random.uniform(0, 20, 200)
salary = 60000 + (experience * 5000) + np.random.normal(0, experience*1000 + 5000, 200)
axes[1, 1].scatter(experience, salary, color='teal', alpha=0.5, s=30)
axes[1, 1].set(title='4. Pay Equity: Experience vs. Salary', xlabel='Years of Experience', ylabel='Salary ($)')

# 5. Distance vs. Delivery Cost (Clusters based on shipping zones)
distance = np.random.uniform(5, 1000, 150)
base_cost = 5
cost = base_cost + (distance * 0.02) + np.random.normal(0, 2, 150)
axes[2, 0].scatter(distance, cost, color='saddlebrown', alpha=0.6)
axes[2, 0].set(title='5. Supply Chain: Distance vs. Delivery Cost', xlabel='Distance to Destination (Miles)', ylabel='Delivery Cost ($)')

fig.delaxes(axes[2, 1])
plt.tight_layout()
plt.show()
```

4. Box Plots: Analyzing Spread and Outliers
Use these to evaluate vendor reliability, segment behavior, and process quality control.

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 15))
fig.suptitle('4. Box Plots: Spreads & Outliers', fontsize=18, fontweight='bold')

# 1. Delivery Times by Carrier
np.random.seed(10)
fedex = np.random.normal(3, 0.5, 100)
ups = np.random.normal(3.2, 0.8, 100)
usps = np.random.normal(4, 1.5, 100) # Higher variance/outliers
axes[0, 0].boxplot([fedex, ups, usps], labels=['Carrier A', 'Carrier B', 'Carrier C'], patch_artist=True, boxprops=dict(facecolor='lightblue'))
axes[0, 0].set(title='1. Vendor Reliability: Delivery Days Spread', ylabel='Days to Deliver')

# 2. Quota Attainment by Region
north = np.random.normal(105, 15, 50)
south = np.random.normal(90, 25, 50)
west = np.random.normal(110, 10, 50)
sns.boxplot(data=[north, south, west], ax=axes[0, 1], palette="Set2")
axes[0, 1].set_xticklabels(['North', 'South', 'West'])
axes[0, 1].axhline(100, color='red', linestyle='--', label='100% Quota')
axes[0, 1].legend()
axes[0, 1].set(title='2. Sales Performance: Quota Attainment %', ylabel='% to Goal')

# 3. Customer Lifetime Value (CLV) by Channel (Showing "Whale" outliers)
seo_clv = np.random.lognormal(mean=5, sigma=0.8, size=200)
ads_clv = np.random.lognormal(mean=5.2, sigma=0.5, size=200)
social_clv = np.random.lognormal(mean=4.5, sigma=1.0, size=200) # High spread, some massive whales
axes[1, 0].boxplot([seo_clv, ads_clv, social_clv], labels=['SEO', 'Paid Ads', 'Social'], showfliers=True)
axes[1, 0].set(title='3. High-Value Users: CLV by Channel', ylabel='Lifetime Value ($)')
axes[1, 0].set_yscale('log') # Log scale because of massive outliers

# 4. Defect Rates by Factory Shift (Notched for statistical medians)
shift_1 = np.random.normal(2.1, 0.3, 30)
shift_2 = np.random.normal(2.2, 0.4, 30)
shift_3 = np.random.normal(3.5, 0.8, 30) # Night shift has issues
axes[1, 1].boxplot([shift_1, shift_2, shift_3], labels=['Morning', 'Evening', 'Night'], notch=True, patch_artist=True, boxprops=dict(facecolor='lightcoral'))
axes[1, 1].set(title='4. Quality Control: Defect % by Shift', ylabel='Defect Rate (%)')

# 5. Salary Distributions by Department (Horizontal)
eng = np.random.normal(120000, 20000, 100)
sales = np.random.normal(90000, 35000, 100) # Higher variance due to commission
marketing = np.random.normal(85000, 15000, 100)
axes[2, 0].boxplot([eng, sales, marketing], labels=['Engineering', 'Sales', 'Marketing'], vert=False, patch_artist=True, boxprops=dict(facecolor='mediumaquamarine'))
axes[2, 0].set(title='5. HR Analytics: Salary Spread by Dept', xlabel='Annual Salary ($)')

fig.delaxes(axes[2, 1])
plt.tight_layout()
plt.show()
```

Use these to monitor growth trajectories, seasonality, and the impact of business interventions.

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 15))
fig.suptitle('5. Line Plots: Time Series & Trends', fontsize=18, fontweight='bold')

months = pd.date_range(start='2023-01-01', periods=24, freq='MS')
x_indices = np.arange(24)

# 1. Monthly Active Users (Growth with slight seasonality)
base_mau = 10000
trend = x_indices * 500
seasonality = np.sin(x_indices * (2 * np.pi / 12)) * 2000
mau = base_mau + trend + seasonality + np.random.normal(0, 500, 24)
axes[0, 0].plot(months, mau, marker='o', color='indigo', linewidth=2)
axes[0, 0].set(title='1. Product Adoption: MAU Growth', ylabel='Active Users')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Daily Cash Flow Balance (Volatile)
days = pd.date_range(start='2024-01-01', periods=90, freq='D')
daily_change = np.random.normal(0, 5000, 90)
daily_change[14::15] += 25000 # Invoices paid twice a month
daily_change[29::30] -= 40000 # Payroll hits monthly
cash_balance = 100000 + np.cumsum(daily_change)
axes[0, 1].plot(days, cash_balance, color='green')
axes[0, 1].fill_between(days, cash_balance, 0, where=(cash_balance < 50000), color='red', alpha=0.3, label='Danger Zone (<$50k)')
axes[0, 1].set(title='2. Financial Health: 90-Day Cash Flow', ylabel='Cash on Hand ($)')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].legend()

# 3. Website Conversion Rate (Step increase after UI update)
conv_rate = np.random.normal(2.5, 0.2, 24)
conv_rate[12:] += 1.5 # UI update at month 12 improves conversion
axes[1, 0].plot(months, conv_rate, linestyle='-', marker='s', color='darkorange')
axes[1, 0].axvline(months[12], color='black', linestyle=':', label='New Checkout UI Released')
axes[1, 0].set(title='3. UX Impact: Conversion Rate', ylabel='Conversion %')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend()

# 4. Seasonal Inventory Demand (Clear Sine wave)
demand = 500 + np.sin(x_indices * (2 * np.pi / 12)) * 300 + np.random.normal(0, 30, 24)
axes[1, 1].plot(months, demand, color='dodgerblue', linewidth=2)
axes[1, 1].fill_between(months, demand, alpha=0.2, color='dodgerblue')
axes[1, 1].set(title='4. Supply Chain: Seasonal Product Demand', ylabel='Units Ordered')
axes[1, 1].tick_params(axis='x', rotation=45)

# 5. Customer Churn Rate Trajectory (Downward trend)
churn_start = 8.0
churn_rate = churn_start - (x_indices * 0.2) + np.random.normal(0, 0.3, 24)
churn_rate = np.clip(churn_rate, 1.0, 10.0) # Floor it at 1%
axes[2, 0].plot(months, churn_rate, color='crimson', marker='v')
axes[2, 0].set(title='5. Retention: Monthly Churn Rate', ylabel='Churn %')
axes[2, 0].tick_params(axis='x', rotation=45)

fig.delaxes(axes[2, 1])
plt.tight_layout()
plt.show()
```

**Business Use Cases for Mathematical Curves**

1. The Product Lifecycle (S-Curve / Logistic Function)Business Value: Strategic planning. Products rarely grow linearly. The S-Curve models the slow "Introduction" phase, the rapid "Growth" phase, and the plateauing "Maturity" phase. Knowing where a product sits on this curve dictates whether you should invest heavily in marketing or prepare for its end-of-life.

2. Price Optimization & Profit Maximization (Parabolic Curve)Business Value: Finding the "sweet spot." If you price a product too low, you sell a lot but make no margin. If you price it too high, the margin is great but volume drops to zero. A profit curve (often an inverted U-shape) visualizes the exact price point that maximizes total profit.

3. Customer Survival / Retention (Exponential Decay Curve)Business Value: Lifetime Value (LTV) forecasting. Instead of a simple churn rate, a survival curve models the exact probability that a customer will remain active after $X$ continuous months. It highlights the critical drop-off periods where you need to send retention offers.

4. Quality Control & Tolerances (Bell Curve / Normal Distribution)Business Value: Six Sigma and manufacturing efficiency. By plotting the ideal continuous distribution of a part's size or weight, and marking the acceptable upper and lower specification limits, you can predict exactly what percentage of a manufacturing run will be defective.

5. The Learning / Experience Curve (Logarithmic Decay)Business Value: Cost forecasting. In manufacturing and operations, as cumulative production volume doubles, the cost per unit typically decreases by a predictable percentage (due to workers getting faster and processes getting optimized). This curve proves the financial value of scaling up.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Set professional theme
sns.set_theme(style="whitegrid", palette="muted")
np.random.seed(42) # Ensure reproducible "random" noise

fig, axes = plt.subplots(5, 2, figsize=(16, 26))
fig.suptitle('Mathematical Models vs. Noisy Real-World Data', fontsize=20, fontweight='bold', y=0.98)

# -------------------------------------------------------------------
# 1. Product Lifecycle (S-Curve / Logistic Function)
# -------------------------------------------------------------------
L, k, x0 = 100000, 1.2, 5  # Max adoption, growth rate, inflection point

# Theoretical
x_smooth_1 = np.linspace(0, 10, 500)
y_smooth_1 = L / (1 + np.exp(-k * (x_smooth_1 - x0)))
axes[0, 0].plot(x_smooth_1, y_smooth_1, color='darkmagenta', linewidth=3)
axes[0, 0].set(title='Theory: Ideal Product Lifecycle', xlabel='Years Since Launch', ylabel='Active Users')

# Real-World (Quarterly data over 10 years = 40 points)
x_actual_1 = np.linspace(0.5, 9.5, 40)
y_actual_1 = L / (1 + np.exp(-k * (x_actual_1 - x0))) + np.random.normal(0, 5000, 40)
axes[0, 1].plot(x_smooth_1, y_smooth_1, color='gray', linestyle='--', linewidth=2, label='Theoretical Model')
axes[0, 1].scatter(x_actual_1, y_actual_1, color='darkmagenta', s=60, alpha=0.8, edgecolor='black', label='Actual Quarterly Data')
axes[0, 1].set(title='Reality: Actual User Growth', xlabel='Years Since Launch', ylabel='Active Users')
axes[0, 1].legend()

# -------------------------------------------------------------------
# 2. Price Optimization (Parabolic / Quadratic Curve)
# -------------------------------------------------------------------
# Theoretical Profit Model: y = -2*(x-80)^2 + 7200
x_smooth_2 = np.linspace(10, 150, 500)
y_smooth_2 = -2 * (x_smooth_2 - 80)**2 + 7200
axes[1, 0].plot(x_smooth_2, y_smooth_2, color='forestgreen', linewidth=3)
axes[1, 0].axvline(80, color='red', linestyle=':', label='Optimal Price ($80)')
axes[1, 0].set(title='Theory: Expected Profit per Price Point', xlabel='Selling Price ($)', ylabel='Expected Profit ($)')
axes[1, 0].legend()

# Real-World (Testing 30 random price points)
x_actual_2 = np.random.uniform(20, 140, 40)
y_actual_2 = -2 * (x_actual_2 - 80)**2 + 7200 + np.random.normal(0, 800, 40)
axes[1, 1].plot(x_smooth_2, y_smooth_2, color='gray', linestyle='--', linewidth=2, label='Profit Model')
axes[1, 1].scatter(x_actual_2, y_actual_2, color='forestgreen', s=60, alpha=0.8, edgecolor='black', label='Actual Test Markets')
axes[1, 1].set(title='Reality: A/B Testing Price Points', xlabel='Selling Price ($)', ylabel='Actual Profit ($)')
axes[1, 1].legend()

# -------------------------------------------------------------------
# 3. Customer Survival (Exponential Decay Curve)
# -------------------------------------------------------------------
decay_rate = 0.15

# Theoretical
x_smooth_3 = np.linspace(0, 36, 500)
y_smooth_3 = 100 * np.exp(-decay_rate * x_smooth_3)
axes[2, 0].plot(x_smooth_3, y_smooth_3, color='crimson', linewidth=3)
axes[2, 0].set(title='Theory: Expected Customer Retention', xlabel='Months Since Signup', ylabel='% Retained')

# Real-World (Tracking 20 different monthly cohorts)
x_actual_3 = np.random.uniform(1, 35, 50)
y_actual_3 = 100 * np.exp(-decay_rate * x_actual_3) + np.random.normal(0, 4, 50)
y_actual_3 = np.clip(y_actual_3, 0, 100) # Ensure it doesn't go above 100% or below 0%
axes[2, 1].plot(x_smooth_3, y_smooth_3, color='gray', linestyle='--', linewidth=2, label='Survival Model')
axes[2, 1].scatter(x_actual_3, y_actual_3, color='crimson', s=60, alpha=0.8, edgecolor='black', label='Cohort Data')
axes[2, 1].set(title='Reality: Actual Cohort Retention', xlabel='Months Since Signup', ylabel='% Retained')
axes[2, 1].legend()

# -------------------------------------------------------------------
# 4. Quality Control (Normal Distribution vs Histogram)
# -------------------------------------------------------------------
mean_w, std_w = 100, 1.2

# Theoretical
x_smooth_4 = np.linspace(95, 105, 500)
y_smooth_4 = stats.norm.pdf(x_smooth_4, mean_w, std_w)
axes[3, 0].plot(x_smooth_4, y_smooth_4, color='royalblue', linewidth=3)
axes[3, 0].set(title='Theory: Ideal Manufacturing Tolerances', xlabel='Product Weight (g)', ylabel='Probability Density')

# Real-World (Instead of scatter, we use a histogram of actual measured weights)
actual_weights = np.random.normal(mean_w, std_w + 0.3, 1000) # Slightly higher real-world variance
axes[3, 1].plot(x_smooth_4, y_smooth_4, color='black', linestyle='--', linewidth=3, label='Target PDF')
axes[3, 1].hist(actual_weights, bins=30, density=True, color='royalblue', alpha=0.6, edgecolor='black', label='Actual Factory Output')
axes[3, 1].set(title='Reality: Measured Factory Weights', xlabel='Product Weight (g)', ylabel='Density')
axes[3, 1].legend()

# -------------------------------------------------------------------
# 5. The Learning Curve (Logarithmic Decay)
# -------------------------------------------------------------------
initial_cost, learning_rate = 500, -0.3

# Theoretical
x_smooth_5 = np.linspace(100, 10000, 500)
y_smooth_5 = initial_cost * (x_smooth_5 / 100)**learning_rate
axes[4, 0].plot(x_smooth_5, y_smooth_5, color='darkorange', linewidth=3)
axes[4, 0].set(title='Theory: Expected Experience Curve', xlabel='Cumulative Units Produced', ylabel='Cost Per Unit ($)')

# Real-World (Cost reports at various production milestones)
x_actual_5 = np.linspace(200, 9500, 30)
y_actual_5 = initial_cost * (x_actual_5 / 100)**learning_rate + np.random.normal(0, 15, 30)
axes[4, 1].plot(x_smooth_5, y_smooth_5, color='gray', linestyle='--', linewidth=2, label='Theoretical Cost')
axes[4, 1].scatter(x_actual_5, y_actual_5, color='darkorange', s=60, alpha=0.8, edgecolor='black', label='Actual Batch Costs')
axes[4, 1].set(title='Reality: Actual Unit Costs vs. Volume', xlabel='Cumulative Units Produced', ylabel='Cost Per Unit ($)')
axes[4, 1].legend()

# Clean up layout
plt.tight_layout()
fig.subplots_adjust(top=0.95) # Make room for the main title
plt.show()
```
